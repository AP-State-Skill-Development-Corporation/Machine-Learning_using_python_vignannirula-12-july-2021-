Pandas Series:
------------------------
- A Series is similar to a 1-D numpy array and contains values of the same type.
- A DataFrame is simply a table where each column is a pandas series
- Creating series
    - List
    - Tuple
    - Dictionary
    - Numpy
    - Data_Range

```python
# creating pandas series using list
k = [1,2,3,4,89,76,54,23]
s1 = pd.Series(k)
s1   # index value starts with 0
```

```python
# creating pandas series using tuple
t = (12,34,56,78,90,46,28)
s2 = pd.Series(t)
s2
```

```python
# creating pandas series using dictionary
d = {"a":12,"b":89,"c":520}
s3 = pd.Series(d)
s3
```

```python
s1[3]
```

```python
s2[2]
```

```python
s3["a"]
```

```python
s3.index = [23,45,76] # reindexing the data
s3
```

```python
s3[23]
```
